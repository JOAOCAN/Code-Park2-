```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Criamos uma amostra ficiicia ( Temperatuas em graus)
amostra = np.array([30.6, 36.7, 36.8, 37.0,36.9, 26.6, 36.5, 49.0, 36.7, 36.6])
```

```python
# Estatitiscas descritivas básicas
media = np.mean(amostra)                                                        # média aritmedica
mediana = np.median(amostra)                                                    # Mediana
moda = stats.mode(amostra, keepdims=True) [0] [0]                               #Moda ( Valor mais frequente)
desvio_padrao = np.std(amostra)                                                 # Desvio padrão
```

```python
#Exibindo os valores calculados
print("Amostra:", amostra)
print("Média:", round(media, 2))
print("Mediana:", mediana)
print("Moda:", moda)
print("Desvio padrão:", round(desvio_padrao, 2))
```

```python
# Definindo limites para detecção de anomalias (Outliers)
limite_superior = media + 2 * desvio_padrao
limite_inferior = media - 2 * desvio_padrao
```

```python
#Identificamos os valores fora dos limites
anomalos = amostra[(amostra > limite_superior) | (amostra < limite_inferior)]
print("\nValores anomalos detectados:", anomalos)
```

```python
#geração de graficos para Visualização
plt.figure(figsize=(10, 4))
plt.plot(amostra, 'o-', label='amostra')
```

```python
#Adiciona linha da média
plt.axhline(media, color='blue', linestyle='--', label=f'Média ({round(media, 2)})')
```

```python
#destaque dos pontos considerados anômolos
for i, valor in enumerate(amostra):
  if valor > limite_superior or valor < limite_inferior:
    plt.plot(i, valor, 'ro', label='Anomalia' if i == list(amostra).index(valor) else "" )
```

```python
#personalização do grafico
plt.title('Visualização da amostra com Estatisticas')
plt.xlabel('Índice da Medição')
plt.ylabel('Temperatura (ºC)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
